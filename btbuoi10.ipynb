```python
#w9a3
def find_smallest(a):
  a = str(a)
  n = len(a)
  l = []
  dp = [0]*n
  for i in range(n):
    l.append(int(a[i]))

  dp[n-1] = (l[n-1], n-1)
  for i in range(n-2, -1, -1):
    if l[i] < dp[i+1][0]:
      dp[i] = (l[i], i);
    else:
      dp[i] = dp[i+1]

  swap = False
  for i in range(n-1):
    if l[i] > dp[i+1][0]:
      l[i], l[dp[i+1][1]] = l[dp[i+1][1]], l[i]
      swap = True
      break

  if not swap:
    l[n-1], l[n-2] = l[n-2], l[n-1]

  ans = 0
  for i in l:
    ans = ans*10 + i

  return ans

n = int(input())

print(find_smallest(n))
```

```python
#w9a5
def prefix_func(s):
  n = len(s)
  pi = [0]*n
  for i in range(1, n):
    j = pi[i-1]
    while j>0 and s[i]!=s[j]:
      j = pi[j-1]
    if s[i] == s[j]:
      j+=1
    pi[i] = j
  return pi

def check_subset(l1, l2):
  s = l1 + ['#'] + l2
  pi = prefix_func(s)
  for i in range(len(l1)+1, len(s)):
    if pi[i] == len(l1):
      return 'YES'
  return 'NO'

l1 = [x for x in input().split()]
l2 = [x for x in input().split()]

print(check_subset(l1, l2))
```

```python
#w9a6
import copy

def next_permutation(a):
  n = len(a)
  chosen = [True] * (n+1)
  cur = a[n-1]
  chosen[cur] = False
  for i in range(n-2, -1, -1):
    if a[i]>cur:
      cur = a[i]
      chosen[a[i]] = False
    else:
      old = copy.deepcopy(a[i])
      while chosen[a[i]+1]:
        a[i]+=1
      a[i]+=1
      for j in range(i+1, n):
        if a[j] == a[i]:
          a[j] = old
      for j in range(i+1, i+1+(n-i-1)//2):
        a[j], a[n-1-(j-(i+1))] = a[n-1-(j-(i+1))], a[j]
      return
  for i in range(n//2):
    a[i], a[n-1-i] = a[n-1-i], a[i]

n = int(input())
a = [int(x) for x in input().split()]
next_permutation(a)
for i in a:
  print(i, end=' ')
```

```python
#w9a7
def canPlaceFlowers(flowerbed, k) -> bool:
  pos = [-1]
  for i in range(len(flowerbed)):
    if flowerbed[i] == 1: pos.append(i)
  if len(pos)==1:
    return (len(flowerbed) + 1)//2 >= k
  pos.append(len(flowerbed))
  total = 0
  for i in range(len(pos)-1):
    l = pos[i+1]-pos[i]-1
    if i==0 or i==len(pos)-1: total += l//2
    else: total += max((l-1), 0)//2
  return total >= k

flowerbed = [int(x) for x in input().split()]
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#w9a8
crossed = {(0,0):True}
cur = [0,0]
n = 0
def is_path_crossing(moves):
  global crossed, cur, n
  if n==len(moves):
    n = 0
    return False
  if moves[n]=='R': cur[0]+=1
  if moves[n]=='L': cur[0]-=1
  if moves[n]=='U': cur[1]+=1
  if moves[n]=='D': cur[1]-=1
  if (cur[0], cur[1]) in crossed:
    n = 0
    return True
  crossed[(cur[0], cur[1])] = True
  n+=1
  return is_path_crossing(moves)

s = input()
print(is_path_crossing(s))
```

```python
#w9a10
def findRadius(houses, heaters):
    mi = houses[0]
    ma = houses[0]
    for i in houses:
        mi = min(mi, i)
        ma = max(ma, i)
    ans = 0
    for i in heaters:
        r = max(abs(i - mi), abs(i-ma))
        ans = max(ans, r)
    return ans

houses = [int(x) for x in input().split()]
heaters = [int(x) for x in input().split()]

print(findRadius(houses, heaters))
```
