```python
#w8a3
def square_digit_sum(n:int)->int:
  ans = 0
  while n:
    ans += (n%10) * (n%10)
    n//=10
  return ans

def is_happy(n:int) -> int:
  exist = {}
  exist[n] = True
  while n!=1:
    n = square_digit_sum(n)
    if n in exist:
      return 'NO'
    exist[n] = True
  return 'YES'

a = 0

while True:
  try:
    a = int(input())
    break
  except ValueError:
    pass

print(is_happy(a))
```

```python
#w10a7
s = input()
txt = '.txt'
zip = '.zip'
is_txt = True
is_zip = True
try:
  for i in range(-1, -5, -1):
    if s[i].lower() != txt[i]:
      is_txt = False
      break
  for i in range(-1, -5, -1):
    if s[i].lower() != zip[i]:
      is_zip = False
      break
  if is_txt or is_zip:
    print('Doc file thanh cong')
  else:
    print('File khong hop le')
except IndexError:
    print('File khong hop le')
```
