```python
#w11a4
def areAnagrams(path1: str, path2: str):
    s1 = ''
    s2 = ''
    with open(path1, 'r') as f:
        s1 = f.read()
    with open(path2, 'r') as f:
        s2 = f.read()
    lst1 = [0]*130
    lst2 = [0]*130
    for i in s1:
        lst1[ord(i.lower())]+=1
    for i in s2:
        lst2[ord(i.lower())]+=1
    for i in range(130):
        if i!=32 and lst1[i] != lst2[i]:
            return False
    return True

inp1 = input()
inp2 = input()

print(areAnagrams(inp1, inp2))
```

```python
#w11a5
def count(path:str, character:str):
    s = ''
    with open(path, 'r') as f:
        s = f.read()
    ans = 0
    for i in s:
        if i==character:
            ans+=1
    return ans

inp = input()
char = input()


print(count(inp, char))
```

```python
#w11a6
def countCharacter(path: str, letter: str):
    s = ''
    with open(path, 'r') as f:
        s = f.read()
    ans = 0
    for i in s:
        if i == letter:
            ans += 1
    return ans

inp = input()
letter = input()
print(countCharacter(inp, letter))
```

```python
#w11a10
def getMoney(path: str) -> list[int]:
    lst = []
    with open(path, 'r') as f:
        s = f.read()
        lst = s.split()
        for i in range(len(lst)):
            lst[i] = int(lst[i])
    return lst

def rob(houses: list[int]) -> int:
    n = len(houses)
    rob1 = 0
    for i in range(0, n, 2):
        rob1 += houses[i]
    rob2 = 0
    for i in range(1, n, 2):
        rob2 += houses[i]
    ans = max(rob1, rob2)
    return ans

inp = input()
houses = getMoney(inp)
print(rob(houses))
```

```python
#w11a12
def maximumProduct (path:str) -> int:
    ans = -999999999
    with open(path, 'r') as f:
        lines = f.readlines()
        a = [int(x) for x in lines[1].split()]
        for i in range(len(a)):
            for j in range(len(a)):
                for k in range(len(a)):
                    if i!=j and i!=k and j!=k:
                        ans = max(ans, a[i] * a[j] * a[k])
    return ans

inp = input()
print(maximumProduct(inp))
```

```python
#w1113
def moveZeroes(fileName: str) -> list:
    ans = []
    cnt = 0
    with open(fileName, 'r') as f:
        lines = f.readlines
        a = [int(x) for x in lines[1].split()]
        for i in a:
            if i!=0:
                ans.append(i)
            else:
                cnt+=1
    for i in range(cnt):
        ans.append(0)
    return ans

inp = input()
print(moveZeroes(inp))
```

```python
#w11a17
def productExceptSelf(path: str)->list[int]:
    ans = []
    with open(path, 'r') as f:
        lines = f.readlines()
        a = [int(x) for x in lines[1].split()]
        for i in range(len(a)):
            cur = 1
            for j in range(len(a)):
                if i!=j:
                    cur*=a[j]
            ans.append(cur)
    return ans

inp = input()
ans = productExceptSelf(inp)
for i in ans:
    print(i, end=' ')
```

```python
#w11a18
import math

def estimatedTime(path:str):
    src = [0,0]
    dest = [0,0]
    vel = 0
    with open(path, 'r') as f:
        lines = f.readlines()
        l0 = lines[0].split()
        l1 = lines[1].split()
        l2 = lines[2].split()
        src[0] = int(l0[2])
        src[1] = int(l0[4])
        dest[0] = int(l1[2])
        dest[1] = int(l1[4])
        vel = int(l2[1])
    dist = math.sqrt((dest[0]-src[0])**2 + (dest[1]-src[1])**2)
    ans = dist/vel
    return f"{ans:.2f}"


path = input()
print(estimatedTime(path))
```

```python
#w11a20
def imageSmoother(path:str):
    img = []
    with open(path, 'r') as f:
        img = f.readlines()
    for i in range(len(img)):
        img[i] = img[i].split()
        for j in range(len(img[i])):
            img[i][j] = int(img[i][j])
    m = img[0][0]
    res = [[0]*m for _ in range(m)]
    for i in range(1, m+1):
        for j in range(m):
            s = 0
            cnt = 0
            for x in [-1,0,1]:
                for y in [-1, 0, 1]:
                    if 1<=i+x<=m and 0<=j+y<m:
                        s+=img[i+x][j+y]
                        cnt+=1
            res[i-1][j] = s//cnt
    print(img)
    return res
            

path = input()
print(imageSmoother(path))
```
