```python
def Max_Of_Two(a, b) -> int:
    if a>b:
        return a
    else:
        return b
    
a,b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
def swap(a, b) -> tuple[int, int]:
    a, b = b, a
    return a, b

a, b = map(int, input().split())
print(swap(a, b))
```

```python
from math import sqrt

def prime(a:int)->int:
    if a==0 or a==1:
        return False
    sqrta = int(sqrt(a))+1
    for i in range(2, sqrta):
        if a%i==0:
            return False
    return True

a = int(input())
print(prime(a))
```

```python
from math import sqrt

def divsum(n:int)->int:
    ans=0
    sqrtn = int(sqrt(n))+1
    for i in range(2, sqrtn):
        if n%i==0:
            ans+=i+(n//i)
    ans+=1
    return ans

def perfect(n):
    return n==divsum(n)

a = int(input())
print(perfect(a))
```

```python
def find(l:list[int], k:int)->int:
    for idx, i in enumerate(l, start=1):
        if i==k:
            return idx
    return -1

a = [int(x) for x in input().split()]
k = int(input())
print(find(a, k))
```

```python
def fact(n:int)->int:
    ans=1
    for i in range(2, n+1):
        ans*=i
    return ans

a = int(input())
print(fact(a))
```

```python
def calc(n1:float, n2:float, op:str)->int:
    ans = 0.00
    if op=="+":
        ans = n1+n2
    elif op=="-":
        ans = n1-n2
    elif op=="*":
        ans = n1*n2
    else:
        ans = n1/n2
    return ans

n1, op, n2 = input().split()
n1 = float(n1)
n2 = float(n2)
print(f"{calc(n1, n2, op):.2f}")
```

```python
def hamming(a:int, b:int)->int:
    ans=0
    h=a^b
    while h:
        if h&1:
            ans+=1
        h>>=1
    return ans

a, b = map(int, input().split())
print(hamming(a, b))
```

```python
def sum(n:int)->int:
    ans =0
    while n:
        ans+=n%10
        n//=10
    return ans

a = int(input())
print(sum(a))
```

```python
from collections import defaultdict

def mapable(s1:str, s2:str)->bool:
    s2map = defaultdict(str)
    s1map = defaultdict(str)
    for idx, i in enumerate(s1):
        if s1map[i]=="":
            s1map[i]=s2[idx]
        elif s1map[i]!=s2[idx]:
            return False
    for idx, i in enumerate(s2):
        if s2map[i]=="":
            s2map[i]=s1[idx]
        elif s2map[i]!=s1[idx]:
            return False
    return True

s1, s2 = input().split()
print(mapable(s1, s2))
```
